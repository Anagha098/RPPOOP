```python
class BankAccount:
    def __init__(self, name, accNum, accType):
        self.name = name
        self.accNum = accNum
        self.balance = 0
        self.accType = accType
        
    def Deposit(self, amount):
        self.balance = self.balance + amount
        
    def Withdrawl(self, amount):
        if self.balance >= amount:
            self.balance = self.balance - amount
            print("Amount withdrawl successfull!!!")
        else:
            print("Insufficient balance!!!")
    
    def Display(self):
        print("Name : ",self.name)
        print("Account Number : ",self.accNum)
        print("Account Type : ",self.accType)
        print("Balance : ",self.balance)

#Menu Driven Program
name = input("Enter your name : ")
num = input("Enter your bank account number : ")
Type = input("Enter your bank account type : ")

my_acc = BankAccount(name, num, Type)

while 1>0 :
    print("Enter 1 for deposition \nEnter 2 for withdrawl \nEnter 3 for Account details \nEnter 4 to exit")
    print("\n")
    choice = input("Enter your choice : ")
    print("\n")
          
    
    if choice == "1":
        amount = float(input("Enter the ammount to deposit : "))
        my_acc.Deposit(amount)
        print("Amount deposited successfully!!!")
        print("\n")
    elif choice == "2":
        amount = float(input("Enter the amount to withdraw : "))
        my_acc.Withdrawl(amount)
        
        print("\n")
    elif choice == "3":
        my_acc.Display()
        print("\n")
    elif choice == "4":
        print("Exiting the system!!!")
        print("\n")
        break
    else:
        print("Invalid entry!!!")
        print("\n")
    
```
